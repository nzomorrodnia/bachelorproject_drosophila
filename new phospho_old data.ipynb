```python
all_genes_p = df_chol_prot_data[['Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9']].dropna()
all_genes_p = all_genes_p.to_numpy()[1:]

all_genes_c = df_chol_prot_data[['Unnamed: 19', 'Unnamed: 20', 'Unnamed: 21', 'Unnamed: 22', 'Unnamed: 23', 'Unnamed: 24', 'Unnamed: 25']].dropna()
all_genes_c = all_genes_c.to_numpy()[1:]

plt.plot(all_genes_p.mean(axis=0))
plt.plot(all_genes_c.mean(axis=0))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming all_genes is a numpy array containing the gene expression data

fig, ax = plt.subplots(1,1,figsize=(12,6))

mean_labels = ['Mean Protein diet', 'Mean cholesterol diet']
std_labels = ['Standard Deviation Protein diet', 'Standard Deviation Cholesterol diet']

for i, genes in enumerate([all_genes_p, all_genes_c]):

    # Calculating the mean and standard error
    mean_values = genes.mean(axis=0)
    std_errors = genes.var(axis=0) / np.sqrt(genes.shape[0])  # Standard error = standard deviation / sqrt(sample size)

    # Setting the confidence level (e.g., 95%)
    confidence_level = 0.95

    # Calculating the margin of error (z-score * standard error)
    z_score = 1.96  # For a 95% confidence level
    margin_of_error = z_score * std_errors

    # Calculating the lower and upper bound of the confidence interval
    lower_bounds = mean_values - margin_of_error
    upper_bounds = mean_values + margin_of_error

    lower_bounds = lower_bounds.astype(float)
    upper_bounds = upper_bounds.astype(float)

    ax.plot(mean_values, label=mean_labels[i])
    ax.fill_between(range(len(mean_values)), lower_bounds, upper_bounds, alpha=0.3, label=std_labels[i])

    # Rest of the code

    # Adding labels and title to the plot
    ax.set_xlabel('Time')
    ax.set_ylabel('Abundance Ratio (Phosphorylation level)')
    ax.set_title('Protein phosphorylation over time with Confidence Intervals')

    xlabels = ['15 min.', '30 min.', '1 hour', '2 hours', '4 hours', '6 hours', '10 hours']
    ax.set_xticks([0,1,2,3,4,5,6])
    ax.set_xticklabels(xlabels)

    # Adding a legend
    ax.legend()

# Displaying the plot
plt.show()
```
