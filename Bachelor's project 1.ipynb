```python
PATH2 = '/Users/ninazomorrodnia/Desktop/Bachelorprojekt/Phosphoproteomics Data/Cholesterol re-feeding'
FILE2 = PATH2 + '/Cholesterol refeeding Phospho proteomics volcano plot calculations.xlsx'

# Reading in my data as a Pandas DataFrame for volcano plot
df2 = pd.read_excel(FILE2)
```

```python
df2.head()
```

```python
# Defining the threshold for statistical significance
p_value_threshold = 0.05
fold_change_threshold = 1.5
# The fold change is the difference in expression levels between the ratio
# of expression levels in one condition (feeded cholesterol) relative to
# another (without cholesterol feeded). The fold change threshold is a
# cutoff value used to define what level of change is considered significant.
# Here it means that only changes in expression greater than 1.5-fold will be
# considered, be it positive or negative.
# The 'fold_change_threshold' is used to define the threshold for statistical
# significance for the log2(fold change) values on the x-axis.
```

```python
# Creating new column in DataFrame to indicate whether each row meets the threshold criteria
df2['significance'] = (df2['Abundance Ratio P-Value: (T15min) / (T0)'] < p_value_threshold) & (abs(df2['Log2 Fold Change']) > fold_change_threshold)
```

```python
# Defining colors for each point in the plot based on whether it is significant or not
colors_cholesterol = ['red' if s else 'blue' for s in df2['significance']]

# Here, any log2(fold change) value greater than or equal to 'fold_change_threshold' or less than or equal to negative 'fold_change_threshold'
# will be considered significant and plotted in red. Any log2(fold change) value between the negative and positive 'fold_change_threshold'
# will be plotted in blue.
```

```python
df2.head()
```

```python
# Creating the plots
plt.scatter(df2['Log2 Fold Change'], -1 * df2['Neg Log10 (p-value)'], color=colors_cholesterol, alpha=0.5)
plt.axvline(x=-fold_change_threshold, linestyle='--', color='grey')
plt.axvline(x=fold_change_threshold, linestyle='--', color='grey')
plt.axhline(y=-1* np.log10(p_value_threshold), linestyle='--', color='grey')
plt.title('Volcano plot of cholesterol reefeeding')
plt.xlabel('Log2 Fold Change')
plt.ylabel('Neg Log10 (p-value)')
plt.show
```
